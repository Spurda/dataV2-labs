#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
variables= ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']
for y in variables:
    plt.scatter(fitbit['Steps'], fitbit[y])
```

```python
'''# your code here fig, [ax1, ax2] = plt.subplots(1,2, figsize=(15,5))
variables = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']
for y in variables:
    fig, [ax1, ax2, ax3] = plt.scatter(fitbit['Steps'], fitbit[y])
    ax1.plot(fitbit['Steps'],y)
    ax2.plot(fitbit['Steps'],y)
    ax3.plot(fitbit['Steps'],y)
    plt.show()
for pos, axis1 in enumerate(variables):   # Pick a first col
    for axis2 in enumerate(variables[pos+1:]):   # Pick a later col
        plt.scatter(fitbit.loc[:, axis1], fitbit.loc[:, axis2])'''
```

```python
# 3 separated graphs
variables = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']
for y in variables:

    figure = plt.figure
    ax = plt.gca()
    ax.scatter(fitbit['Steps'], fitbit[y])
    ax.set_xlabel('Steps')
    ax.set_ylabel(y)
    ax.set_title("{} vs {}".format('Steps', y))

    plt.show()
```

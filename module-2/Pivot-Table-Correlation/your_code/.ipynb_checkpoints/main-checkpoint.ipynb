#### What's the mean `Salary` by `Department`?
**Tip**: There are 2 ways to do it: with or without pivot tables. Do it both ways. 

```python
# Method 1
# your code here
employees.pivot_table(values=['Salary'], index=['Department'])
```

```python
employees.pivot_table(columns=['Department'],values=['Salary']).mean()
```

```python
# Method 2
# your code here
employees.groupby(['Department'])[['Salary']].mean()
```

#### Now, group the data by more than one column. Use Pandas pivot table function to select the mean `Salary` by `Department` and `Title`.

```python
# your code here
employees.pivot_table(values=['Salary'], index=['Department', 'Title'])
```

#### Display the number of `Years` of work experience and the `Salary` for each `Title`. Add `Years` as a column in your pivot table.
**Hint**: Use Pandas documentation to explore the [pivot table](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) function.

```python
# your code here
employees.pivot_table(values=['Salary'], index=['Department','Title'], columns=['Years'])
```

#### Find the mean `Salary` and number of employees in each `Department` and for each job `Title`.

```python
# your code here
employees.pivot_table(index=['Department'], values=['Salary'], columns=['Title'])
```

```python
employees.pivot_table(index=['Department'], columns=['Title'], values=['Name'], aggfunc='count')
```

```python
employees.pivot_table(index=['Title','Department'], values=['Name', 'Salary'], aggfunc={'Name':'count', 
                                                                                                   'Salary':'mean'})
```

#### The stake holders want to know, for each `Department`, the number of employees and how much money is spent on salaries. Could you provide that information? 

```python
# your code here
employees.pivot_table(index=['Department'], values=['Salary','Name'], aggfunc={'Salary':'sum',
                                                                              'Name':'count'})
```
